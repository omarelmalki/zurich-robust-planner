```python
import sys
MODULES_PATH = "../../modules"
if MODULES_PATH not in sys.path:
    sys.path.append(MODULES_PATH)
    
from hive_wrapper import HiveConnection
from utils import get_env_vars
```

```python
username, _, _ = get_env_vars()
hive_connection = HiveConnection(*get_env_vars())
```

```python
# all edges table creation
queries = [
    # drop mean tables
    f"""
    DROP TABLE IF EXISTS {username}.avg_delay_tsptype_tspsubtype_dper
    """,
    f"""
    DROP TABLE IF EXISTS {username}.avg_delay_tsptype_dper
    """,
    f"""
    DROP TABLE IF EXISTS {username}.avg_delay_dper
    """,
    f"""
    DROP TABLE IF EXISTS {username}.avg_delay
    """,
    
    # drop std tables
    f"""
    DROP TABLE IF EXISTS {username}.std_delay_tsptype_tspsubtype_dper
    """,
    f"""
    DROP TABLE IF EXISTS {username}.std_delay_tsptype_dper
    """,
    f"""
    DROP TABLE IF EXISTS {username}.std_delay_dper
    """,
    f"""
    DROP TABLE IF EXISTS {username}.std_delay
    """,
    
    # create mean tables
    f"""
    CREATE EXTERNAL TABLE {username}.avg_delay_tsptype_tspsubtype_dper(
        stop_name STRING,
        day_period STRING,
        transport_type STRING,
        transport_subtype STRING,
        mean_arrival_delay DOUBLE
    )
    STORED AS PARQUET
    LOCATION '/user/{username}/delays/avg_delay_tsptype_tspsubtype_dper'
    """,
    f"""
    CREATE EXTERNAL TABLE {username}.avg_delay_tsptype_dper(
        stop_name STRING,
        day_period STRING,
        transport_type STRING,
        mean_arrival_delay DOUBLE
    )
    STORED AS PARQUET
    LOCATION '/user/{username}/delays/avg_delay_tsptype_dper'
    """,
    f"""
    CREATE EXTERNAL TABLE {username}.avg_delay_dper(
        stop_name STRING,
        day_period STRING,
        mean_arrival_delay DOUBLE
    )
    STORED AS PARQUET
    LOCATION '/user/{username}/delays/avg_delay_dper'
    """,
    f"""
    CREATE EXTERNAL TABLE {username}.avg_delay(
        stop_name STRING,
        mean_arrival_delay DOUBLE
    )
    STORED AS PARQUET
    LOCATION '/user/{username}/delays/avg_delay'
    """,
    
    # create std tables
    f"""
    CREATE EXTERNAL TABLE {username}.std_delay_tsptype_tspsubtype_dper(
        stop_name STRING,
        day_period STRING,
        transport_type STRING,
        transport_subtype STRING,
        std_arrival_delay DOUBLE
    )
    STORED AS PARQUET
    LOCATION '/user/{username}/delays/std_delay_tsptype_tspsubtype_dper'
    """,
    f"""
    CREATE EXTERNAL TABLE {username}.std_delay_tsptype_dper(
        stop_name STRING,
        day_period STRING,
        transport_type STRING,
        std_arrival_delay DOUBLE
    )
    STORED AS PARQUET
    LOCATION '/user/{username}/delays/std_delay_tsptype_dper'
    """,
    f"""
    CREATE EXTERNAL TABLE {username}.std_delay_dper(
        stop_name STRING,
        day_period STRING,
        std_arrival_delay DOUBLE
    )
    STORED AS PARQUET
    LOCATION '/user/{username}/delays/std_delay_dper'
    """,
    f"""
    CREATE EXTERNAL TABLE {username}.std_delay(
        stop_name STRING,
        std_arrival_delay DOUBLE
    )
    STORED AS PARQUET
    LOCATION '/user/{username}/delays/std_delay'
    """
]

for query in queries:
    hive_connection.exec(query)
```
