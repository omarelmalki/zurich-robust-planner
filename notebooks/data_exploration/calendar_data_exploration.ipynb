This file contains the weekly schedule for each service. We can use the SERVICE_ID column along with the day of the week columns (MONDAY, TUESDAY,...) to determine if a trip operates on a specific day. This information is important for determining the availability of trips when planning routes.

```python
# Only processing I did was to convert START_DATE and END_DATE columns to datetime
```

```python
import pandas as pd
```

```python
filepath = "../data/calendar.txt"
df = pd.read_csv(filepath)
```

```python
df.head()
```

```python
df['start_date'] = pd.to_datetime(df['start_date'], format='%Y%m%d')
df['end_date'] = pd.to_datetime(df['end_date'], format='%Y%m%d')
```
